## Tumor Response to Treatment

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
drug_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_data = pd.merge(mouse_data, drug_data, on='Mouse ID')

# Display the data table for preview
combined_data.head()
```

```python
#Tumor Response to Treatment

# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
mean_tumor_volume_df = combined_data.groupby(["Drug", "Timepoint"]) 

# Convert to DataFrame and calculate mean.
mean_tumor_volume_df = pd.DataFrame(mean_tumor_volume_df['Tumor Volume (mm3)'].mean())

# Preview DataFrame
mean_tumor_volume_df = mean_tumor_volume_df.unstack(0)
mean_tumor_volume_df = mean_tumor_volume_df['Tumor Volume (mm3)']
mean_tumor_volume_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
se_tumor_volume_df = combined_data.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].sem()

# Convert to DataFrame
se_tumor_volume_df = pd.DataFrame(se_tumor_volume_df)

# Preview DataFrame
se_tumor_volume_df = se_tumor_volume_df.unstack(0)
se_tumor_volume_df = se_tumor_volume_df['Tumor Volume (mm3)']
se_tumor_volume_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
se_tumor_volume_df.unstack(0)

# Preview that Reformatting worked
se_tumor_volume_df
```

```python
# Generate the Plot (with Error Bars)
x_axis = np.arange(0, 50, 5)
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Tumor Response to Treatment")
plt.grid()

drug_format_list = [('Capomulin','red', 'o'),
                    ('Infubinol', 'blue', '^'),
                    ('Ketapril', 'green', 's'),
                    ('Placebo', 'black', 'd')
                   ]

for drug, colors, marker in drug_format_list:
    plt.errorbar(x_axis, mean_tumor_volume_df[drug], se_tumor_volume_df[drug],
                 marker=marker, color=colors, linewidth=1, label=drug)

plt.legend(loc="upper left")

# Save the Figure
plt.savefig("tumor_response.png")
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_metastatic_df = combined_data.groupby(["Drug", "Timepoint"]) 
mean_metastatic_df

# Convert to DataFrame
mean_metastatic_df = pd.DataFrame(mean_metastatic_df['Metastatic Sites'].mean())
mean_metastatic_df

# Preview DataFrame
mean_metastatic_df = mean_metastatic_df.unstack(0)
mean_metastatic_df = mean_metastatic_df['Metastatic Sites']
mean_metastatic_df
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
se_metastatic_df = combined_data.groupby(['Drug','Timepoint'])['Metastatic Sites'].sem()


# Convert to DataFrame
se_metastatic_df = pd.DataFrame(se_metastatic_df)

# Preview DataFrame
se_metastatic_df = se_metastatic_df.unstack(0)
se_metastatic_df = se_metastatic_df['Metastatic Sites']
se_metastatic_df
```

```python
# Generate the Plot (with Error Bars)
x_axis = np.arange(0, 50, 5)
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")
plt.title("Metastatic Spread During Treatment")
plt.grid()

drug_format_list = [('Capomulin','red', 'o'),
                    ('Infubinol', 'blue', '^'),
                    ('Ketapril', 'green', 's'),
                    ('Placebo', 'black', 'd')
                   ]

for drug, colors, marker in drug_format_list:
    plt.errorbar(x_axis, mean_metastatic_df[drug], se_metastatic_df[drug],
                 marker=marker, color=colors, linewidth=1, label=drug)

plt.legend(loc="upper left")

# Show the Figure
plt.show()

# Save the Figure
plt.savefig("metastatic_response.png")
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_count_df = combined_data.groupby(["Drug", "Timepoint"])

# Convert to DataFrame
mouse_count_df = pd.DataFrame(mouse_count_df['Mouse ID'].count())
mouse_count_df

# Preview DataFrame
mouse_count_df = mouse_count_df.unstack(0)
mouse_count_df = mouse_count_df['Mouse ID']
mouse_count_df
```

```python
# Generate the Plot (Accounting for percentages)
x_axis = np.arange(0, 50, 5)
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%) ")
plt.title("Survival During Treatment")
plt.grid()

drug_format_list = [('Capomulin','red', 'o'),
                    ('Infubinol', 'blue', '^'),
                    ('Ketapril', 'green', 's'),
                    ('Placebo', 'black', 'd')
                   ]

for drug, colors, marker in drug_format_list:
    plt.errorbar(x_axis, mouse_count_df[drug]/mouse_count_df[drug][0] * 100,
                 marker=marker, color=colors, linewidth=1, label=drug)

plt.legend(loc="lower left")

# Show the Figure
plt.savefig("survival_rate.png")
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
tumor_change = (mean_tumor_volume_df.loc[45, :] - mean_tumor_volume_df.loc[0, :]) / mean_tumor_volume_df.loc[0, :] * 100
tumor_change
```

```python
# Store all Relevant Percent Changes into a Tuple

# Splice the data between passing and failing drugs
passing = tumor_change <= 0

drugs_list = [('Capomulin'),
              ('Infubinol'),
              ('Ketapril'),
              ('Placebo')
              ]

# Orient widths. Add labels, tick marks, etc. 
tumor_passing_list = [(tumor_change[x])for x in drugs_list]

barplot = plt.bar(drugs_list, tumor_passing_list, width=-1, align='edge',
        color=passing.map({True:'g',False:'r'})
        )

plt.grid()
plt.ylim(-25, 70)
plt.ylabel("% Tumor Volume Change")
plt.title("Tumor Change Over 45 Day Treatment")


# Use functions to label the percentages of changes
def label_function(labels):
    for label in labels:
        height = label.get_height()
        if height > 0:
            label_position = 2
        else:
            label_position = -7
        plt.text(label.get_x() + label.get_width()/2., label_position,
                '%d' % int(height)+'%',color='white',
                ha='center', va='bottom')


# Call functions to implement the function calls
label_function(barplot)

# Save the Figure
plt.savefig("summary_bargraph.png")

# Show the Figure
plt.show()
```
